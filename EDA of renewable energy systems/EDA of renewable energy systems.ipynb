  <h3 style="color:#006400; margin-top:0;"> About the Dataset</h3>

  <p>
    This dataset contains detailed and structured information about various 
    <b>renewable energy systems</b>, covering technical, economic, environmental, 
    and socio-economic aspects. It is designed to support 
    <b>Exploratory Data Analysis (EDA)</b>, data-driven decision-making, and 
    sustainability-focused research.
  </p>

  <p>
    The primary goal of this dataset is to provide a <b>comprehensive view of renewable 
    energy performance</b>, enabling analysts and researchers to understand how 
    different renewable technologies perform in terms of energy generation, 
    storage efficiency, investment cost, grid integration, and environmental impact.
  </p>

  <h4 style="color:#006400;"> Variables Description</h4>

  <ul>
    <li>
      <b>Type_of_Renewable_Energy:</b> Numerical code representing the renewable energy source 
      (1: Solar, 2: Wind, 3: Hydroelectric, 4: Geothermal, 5: Biomass, 6: Tidal, 7: Wave).
    </li>
    <li>
      <b>Installed_Capacity_MW:</b> Installed power capacity of the system in megawatts (MW).
    </li>
    <li>
      <b>Energy_Production_MWh:</b> Annual energy production measured in megawatt-hours (MWh).
    </li>
    <li>
      <b>Energy_Consumption_MWh:</b> Annual energy consumption in megawatt-hours (MWh).
    </li>
    <li>
      <b>Energy_Storage_Capacity_MWh:</b> Maximum energy storage capacity in megawatt-hours (MWh).
    </li>
    <li>
      <b>Storage_Efficiency_Percentage:</b> Efficiency of the energy storage system (%).
    </li>
    <li>
      <b>Grid_Integration_Level:</b> Level of integration with the electrical grid 
      (1: Fully Integrated, 2: Partially Integrated, 3: Minimal Integration, 4: Isolated Microgrid).
    </li>
    <li>
      <b>Initial_Investment_USD:</b> Initial capital investment required for the system (USD).
    </li>
    <li>
      <b>Funding_Sources:</b> Numerical code representing funding source 
      (1: Government, 2: Private, 3: Public–Private Partnership).
    </li>
    <li>
      <b>Financial_Incentives_USD:</b> Financial incentives or subsidies received (USD).
    </li>
    <li>
      <b>GHG_Emission_Reduction_tCO2e:</b> Reduction in greenhouse gas emissions 
      (tons of CO₂ equivalent).
    </li>
    <li>
      <b>Air_Pollution_Reduction_Index:</b> Index representing air pollution reduction impact.
    </li>
    <li>
      <b>Jobs_Created:</b> Number of jobs generated by the renewable energy project.
    </li>
  </ul>

  <h4 style="color:#006400;"> Aim of Exploratory Data Analysis (EDA)</h4>

  <p>
    The EDA aims to uncover patterns, trends, and relationships within the data to:
  </p>

  <ul>
    <li>Compare performance across different renewable energy technologies</li>
    <li>Analyze the relationship between installed capacity and energy production</li>
    <li>Evaluate how investment and incentives impact environmental benefits</li>
    <li>Assess the role of grid integration and storage efficiency</li>
    <li>Identify outliers, data quality issues, and hidden insights</li>
  </ul>

  <h4 style="color:#006400;"> Problem(s) This Dataset Helps Solve</h4>

  <ul>
    <li>Which renewable energy type provides the highest environmental benefit per investment?</li>
    <li>How does grid integration influence energy efficiency and reliability?</li>
    <li>What factors drive higher GHG emission reduction?</li>
    <li>Which renewable technologies create more jobs relative to capacity?</li>
    <li>How effective are financial incentives in improving sustainability outcomes?</li>
  </ul>

  <p>
    Overall, this dataset is ideal for <b>EDA, sustainability analytics, policy evaluation, 
    and machine learning applications</b> in the renewable energy domain.
  </p>

```python
 #lets map the strin values  
df['Type_of_Renewable_Energy'] = df['Type_of_Renewable_Energy'].astype(str)
 
mapping = {
    '1': 'Solar',
    '2': 'Wind',
    '3': 'Hydroelectric',
    '4': 'Geothermal',
    '5': 'Biomass',
    '6': 'Tidal',
    '7': 'Wave'
}

 
df['Type_of_Renewable_Energy'] = df['Type_of_Renewable_Energy'].map(mapping)
```

```python
 
df['Funding_Sources'] = df['Funding_Sources'].astype(str)
 
funding_mapping = {
    '1': 'Government',
    '2': 'Private',
    '3': 'Public-Private Partnership'
}
 
df['Funding_Sources'] = df['Funding_Sources'].map(funding_mapping)
```

## Storage Efficiency of Different Types of Renewable Energy

  <p>
    The box plot analysis of <b>energy storage efficiency percentages</b> across different 
    renewable energy sources indicates a generally <b>consistent performance range</b>. 
    For most technologies, storage efficiency values are concentrated between 
    <b>60% and 90%</b>, suggesting comparable maturity and effectiveness of storage systems 
    across the renewable energy spectrum.
  </p>

  <p>
    A closer examination shows that <b>Solar energy</b> exhibits a slightly 
    <b>lower median storage efficiency</b> relative to other renewable sources. 
    This may reflect greater variability in storage technologies used for solar systems 
    or the intermittent nature of solar power generation.
  </p>

  <p>
    In contrast, <b>Tidal</b> and <b>Wave energy</b> display a 
    <b>wider spread in storage efficiency values</b>, indicating higher variability. 
    This broader range may be attributed to the emerging and experimental nature of 
    marine energy storage solutions, where technology performance is still evolving.
  </p>

  <p>
    From an EDA standpoint, this insight highlights the importance of considering 
    <b>variability and reliability</b>—not just median efficiency—when comparing storage 
    performance across renewable energy types.

## Correlation Analysis

```python
# Calculate the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Plot the correlation matrix
plt.figure(figsize=(15, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## Emission Reduction and Jobs Created

```python
grouped_df = df.groupby('Type_of_Renewable_Energy').mean(numeric_only=True).reset_index()

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.barplot(x='Type_of_Renewable_Energy', y='GHG_Emission_Reduction_tCO2e', data=grouped_df, palette='viridis')
plt.title('GHG Emission Reduction by Type')
plt.xlabel('Type of Renewable Energy')
plt.ylabel('GHG Emission Reduction (tCO2e)')
plt.xticks(ticks=range(7), labels=['Solar', 'Wind', 'Hydroelectric', 'Geothermal', 'Biomass', 'Tidal', 'Wave'])

plt.subplot(1, 2, 2)
sns.barplot(x='Type_of_Renewable_Energy', y='Jobs_Created', data=grouped_df, palette='viridis')
plt.title('Jobs Created by Type')
plt.xlabel('Type of Renewable Energy')
plt.ylabel('Number of Jobs Created')
plt.xticks(ticks=range(7), labels=['Solar', 'Wind', 'Hydroelectric', 'Geothermal', 'Biomass', 'Tidal', 'Wave'])

plt.tight_layout()
plt.show()
```
